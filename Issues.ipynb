```python
from bson.code import Code

subjects_map = Code('function() {'
                    '    '
                    '    if (!this.subjects) {'
                    '        return;'
                    '    }'
                    ''
                    '    for (i in this.subjects) {'
                    '        emit(this.subjects[i], 1);'
                    '    }'
                    '}')

subjects_reduce = Code('function(key, values) {'
                        ''
                        '    var count = 0;'
                        ''
                        '    for (i in values) {'
                        '        count += values[i];'
                        '    }'
                        '    return count'
                        '}')

result = connection.processing_cache.bills_govtrack.map_reduce(subjects_map, subjects_reduce, "subjects_simple")
```

```python
subs = db['subjects_simple']

s = []
for c in subs.find():
    s.append(c)
    
spd = pandas.DataFrame(s)
```

```python
spd = spd.sort('value', ascending=False)
```

```python
length = 20

spd = spd.head(length)

fig, ax = plt.subplots()

names = []
values = []

for j, k in spd.iterrows():
    names.append(k['_id'])
    values.append(k['value'])
    
plt.bar(range(len(values)), values, align='center')

ind = range(len(names))

ax.set_xticks(ind)
ax.set_xticklabels(names)

plt.xticks(names, names, color='red')
    
plt.show()
```
